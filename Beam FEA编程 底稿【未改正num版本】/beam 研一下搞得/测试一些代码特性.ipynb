# about odeint

```python
#E = 2*10**11    #测试数字使用 Rho = 1 #测试使用
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
Mit = 1
Kit = 506068796
Cit = 0.2
Fit = 45
def diff(y,t):
    yt, theta = y
    return theta , (Fit-Kit*yt-Cit*theta)/Mit

y0 = [0,0]
t = np.linspace(0, 10, 101)
sol = odeint(diff, y0, t) 
plt.plot(t,sol)
plt.legend()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
Mit = 12
Kit = 51
Cit = 0.5
Fit = 11
def diff(y,t):
    yt, theta = y
    return theta , (Fit*np.sin(t)-Kit*yt-Cit*theta)/Mit

y0 = [0,0]
t = np.linspace(0, 10, 101)
sol = odeint(diff, y0, t) 
plt.plot(t,sol)
plt.legend()
```
