# 4.1 Matplotlib 基础

```python
x = np.linspace(0,20,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
siny = y.copy()
```

```python
plt.plot(x,cosy,color='red',linestyle='--',label='cos')
plt.plot(x,siny,'k-.',linewidth=3.7,label='sin')
plt.legend()#显示标签
plt.xlim(0,5)
plt.ylim(-0.5,1)
plt.grid()#网格
plt.show()#显示图像（vscode不需要）
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #正常显示title的中文，但是仅仅坐标轴不需要这句话
plt.rcParams['axes.unicode_minus']=False #title显示中文后 正负号无法显示，此为修复语句

plt.plot(x,cosy,color='#F00000',linestyle='--',marker='*',label='电压U1') #marker原始数据点
plt.plot(x,siny,'k-',linewidth=3.7,label='电压U2')
plt.legend()#显示标签
plt.xlim(0,5)
plt.ylim(-0.5,1)
plt.grid()#网格
plt.xlabel('时间t/s')
plt.ylabel('电压Volt/V')
plt.title('电压变化曲线')
plt.show()#显示图像（vscode不需要）
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False

plt.rcParams['font.size']=10 #改全局字体大小
plt.figure(figsize=(10,5),dpi=100) #图片大小
plt.plot(x,cosy,color='#F00000',linestyle='--',marker='*',label='电压U1') 
plt.plot(x,siny,'k-',linewidth=3.7,label='电压U2')
plt.legend()
plt.xlim(0,5)
plt.ylim(-0.5,1)
plt.grid()
plt.xlabel('时间t/s')
plt.ylabel('电压Volt/V')
plt.title('电压变化曲线',fontsize=15) #字体大小
plt.savefig('电压变化',dpi=100)
plt.show()
```

## 4.2.1柱状图

```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['font.sans-serif']=['SimHei'] #正常显示title的中文，但是仅仅坐标轴不需要这句话
plt.rcParams['axes.unicode_minus']=False #title显示中文后 正负号无法显示，此为修复语句

#开头设置一次即可
```

```python
cities = np.array(['北京市','上海市','天津市','重庆市'])
GDP2018 = np.array([12400,13900,9386,9143])
```

```python
plt.bar(range(4),GDP2018)
plt.title('四个直辖市GDP')
plt.ylabel('GDP')
plt.xticks(range(4),cities)
```

```python
plt.barh(range(4),GDP2018)
plt.title('四个直辖市GDP')
plt.xlabel('GDP')
plt.yticks(range(4),cities)
```

```python
GDP2019 = np.array([13400,14900,9886,9943])
```

```python
H = 0.4
plt.barh(np.arange(4),GDP2018,height=H)
plt.barh(np.arange(4)+H,GDP2019,height=H) #既控制宽度又控制起点
plt.title('四个直辖市GDP')
plt.xlabel('GDP')
plt.yticks(range(4),cities)
```

```python
W=0.4
plt.bar(np.arange(4),GDP2018,width=W,label='2018')
plt.bar(np.arange(4)+W,GDP2019,width=W,label='2019')
plt.title('四个直辖市GDP')
plt.xticks(range(4),cities)
plt.legend()
plt.ylabel('GDP')
```

## 4.2.4面积图

```python
date = ['2020-01','2020-02','2020-03','2020-04']
eat = [10,20,10,40]
drink = [20,40,30,20]
play = [20,20,50,40]

plt.stackplot(date,eat,drink,play,colors=['red','yellow','blue'])
plt.plot([],[],'r',label='吃')
plt.plot([],[],'y',label='喝')
plt.plot([],[],'r',label='玩')
plt.legend()
```

```python
# 折线图plot：观察一个变量的变化趋势
# 散点图scatter：分析两个量之间的关系
# 柱状图：观察或者对比不同量之间的变化
# 直方图：分析一个变量值的分布特征
# 面积图：观察几个量和总的量之前的变化趋势
```
